# Análise de Personalidade de Cliente

## Descrição

A **Análise de Personalidade do Cliente** envolve um exame completo dos perfis ideais de clientes de uma empresa. \
Essa análise facilita um entendimento mais profundo dos clientes, permitindo que as empresas personalizem produtos \
para atender às distintas necessidades, comportamentos e preocupações de vários tipos de clientes.


## Relevância
- Ao conduzir uma Análise de Personalidade do Cliente, as empresas podem refinar seus produtos com base \
nas preferências de segmentos específicos de clientes.

- Em vez de alocar recursos para comercializar um novo produto para todo o banco de dados de clientes,\
as empresas podem identificar os segmentos com maior probabilidade de se interessar pelo produto.

- Posteriormente, esforços de marketing direcionados podem ser direcionados a esses segmentos específicos,\
otimizando a utilização de recursos e aumentando a probabilidade de adoção bem-sucedida do produto.

```python
import pandas as pd

df = pd.read_csv('./customer_segmentation.csv')
df
```

```python
df = df.select_dtypes(include=['number'])
```

## Tratar valores nulos

```python
print(df.isnull().sum().sum()) # 24 null values

print(df.columns[df.isna().any()].tolist()) # Income has 24 null values

# We can fill these null values with the mean, only in the income column
for column in df.columns[df.isnull().any(axis=0)]:
    # Old method
    # df_mean = df[column].fillna(df[column].mean(),inplace=True)
    df_mean = df.fillna({column: df[column].mean()})
    df_median = df.fillna({column: df[column].median()})
```

## Remover coluna ID

```python
df = df.drop(columns=['ID'])
df_mean = df_mean.drop(columns=['ID'])
df_median = df_median.drop(columns=['ID'])
```

## Importar PCA, Kmeans e Matplotlib

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

## PCA em dois componentes para média e mediana

```python
# Inicializar PCA com o número máximo de componentes que o DataFrame possui
pca_mean = PCA()
pca_median = PCA()

# Ajustar e transformar os dados
X_pca_mean = pca_mean.fit_transform(df_mean)
X_pca_median = pca_median.fit_transform(df_median)

# Variância explicada acumulada
explained_variance_mean = np.cumsum(pca_mean.explained_variance_ratio_)
explained_variance_median = np.cumsum(pca_median.explained_variance_ratio_)

# Plotar o gráfico de variância acumulada
plt.figure(figsize=(14, 7))

# Plot para df_mean
plt.subplot(1, 2, 1)
plt.plot(range(1, len(explained_variance_mean) + 1), explained_variance_mean, marker='o')
plt.title('Variância Acumulada - Média')
plt.xlabel('Número de Componentes')
plt.ylabel('Variância Acumulada Explicada')
plt.grid(True)

# Plot para df_median
plt.subplot(1, 2, 2)
plt.plot(range(1, len(explained_variance_median) + 1), explained_variance_median, marker='o')
plt.title('Variância Acumulada - Mediana')
plt.xlabel('Número de Componentes')
plt.ylabel('Variância Acumulada Explicada')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
pca = PCA(n_components=3)
X_pca_mean = pca.fit_transform(df_mean)
X_pca_median = pca.fit_transform(df_median)
```

## KMeans com 2 clusters

### Média

```python
import matplotlib.pyplot as plt
import numpy as np

# Coeficientes dos componentes principais (loadings)
loadings = pca.components_.T  # Transpor para alinhar com as variáveis originais
columns = df_mean.columns  # Nomes das features originais do dataset

# Número de componentes principais a serem plotados
n_components = 3

# Plotar o gráfico de loadings
plt.figure(figsize=(14, 8))

# Largura das barras
bar_width = 0.25
# Posições das barras no eixo X
index = np.arange(len(columns))

# Plotar as barras para cada componente principal
for i in range(n_components):
    plt.bar(index + i * bar_width, loadings[:, i], bar_width, label=f'PC{i+1}', alpha=0.6)

# Configurações de plotagem
plt.xticks(index + (n_components - 1) * bar_width / 2, columns, rotation=45, ha="right")
plt.title("Loadings das Variáveis nos 3 Primeiros Componentes Principais")
plt.xlabel("Variáveis Originais")
plt.ylabel("Contribuição (Loading)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Aplicar o KMeans com 2 clusters
kmeans2 = KMeans(n_clusters=2, random_state=42)
kmeans2.fit(X_pca_mean)
y_kmeans2 = kmeans2.predict(X_pca_mean)

# Plotar os componentes principais e os clusters
plt.figure(figsize=(8,6))
plt.scatter(X_pca_mean[:, 0], X_pca_mean[:, 1], c=y_kmeans2, cmap='viridis', marker='o', edgecolor='k', s=100)
centers = kmeans2.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')  # plotar os centros
plt.title("Clusters após PCA com KMeans")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.show()
```

## Coeficiente de Silhueta
Se os rótulos da verdade básica não forem conhecidos, a avaliação deve ser realizada usando o próprio modelo. O Coeficiente de Silhueta (sklearn.metrics.silhouette_score) é um exemplo de tal avaliação, onde uma pontuação mais alta do Coeficiente de Silhueta se relaciona \
a um modelo com clusters mais bem definidos. O Coeficiente de Silhueta é definido para cada amostra e é composto de duas pontuações:

***a*** : A **distância média** entre uma amostra e todos os outros **pontos na mesma classe**.

***b*** : A **distância média** entre uma amostra e todos os outros pontos no **cluster mais próximo**.

O Coeficiente de Silhueta s para uma única amostra é então dado como:
$$s = \cfrac{b-a}{max(a,b)}$$

## Gráfico do coeficiente de silhueta para média

```python
from sklearn import metrics
```

```python
valores_k = []
ss = []


for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X_pca_mean)
    valores_k.append(k)
    ss.append(metrics.silhouette_score(X_pca_mean, kmeans.labels_))
```

```python
fig, ax = plt.subplots()

ax.plot(valores_k, ss)
plt.show()
```
